```python
import requests
```

爬取台北各車站資訊


```python
# 經觀察後，發現有一個API紀路個線站別，以及站別ID

station_url = "https://web.metro.taipei/apis/metrostationapi/menuline"

data = {"LineID":"0","Lang":"tw"}

response = requests.post(station_url, data=data)


print(response.json())

```

```python
# 接著可以用這個API，透過給定站別ID，取得各站詳細資訊

station_detail_url = "https://web.metro.taipei/apis/metrostationapi/stationdetail"

# 透過第一步的爬蟲，取得的站別SID，將SID給post的data，取得詳細資訊，此處用019測試
data = {"SID":"019","Lang":"tw"}

response = requests.post(station_detail_url, data=data)


print(response.json())
```
